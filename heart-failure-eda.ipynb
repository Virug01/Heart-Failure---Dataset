```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Importing Data

```python
df = pd.read_csv("/kaggle/input/heart-failure-prediction/heart.csv")
df.head()
```

## Analysis of Numeric Data 

```python
sns.distplot(df["Age"],kde = True , hist = True, rug = True, color = "purple")
# sns.histplot(df["Age"],kde = True, color = "purple")
```

## Categorical Data Analysis By Pie Chart

```python
df.groupby("ChestPainType").size().plot(kind = "pie", autopct = "%.1f", figsize = (4,4))
```

```python
df.groupby("RestingECG").size().plot(kind = "pie", autopct = "%.1f", figsize = (4,4))
```

# Violin Plot

```python
sns.violinplot(y = df["Sex"],x = df["HeartDisease"])
```

### The ratio of males of having a heartdisease is more as compare to the females

```python
sns.violinplot(y = df["Age"],x = df["HeartDisease"])
```

## The majority of person having the Heartdisease are lies between 50 to 60 

## Analysis of ChestPain 

```python

p = sns.displot(data=df, x="HeartDisease", hue="ChestPainType", multiple="stack")

plt.xticks([-1, 2])
plt.show()
```

### The ASY ChestPain type of a heart disease person is most common in our data set.
### And the ATA, NAP ChestPain Type of a non Diesease person is common in our data set
